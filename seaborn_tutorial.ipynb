```python
#Importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# Visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot:

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# loadint the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

**Count Plot**

```python
sns.countplot(x='class',data=titanic,hue="class")
```

```python
sns.countplot(x="survived",data=titanic,hue="survived")
```

**Bar Chart**

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
from sklearn.datasets import fetch_california_housing
house_dataset = fetch_california_housing()

house = pd.DataFrame(house_dataset.data,columns=house_dataset.feature_names)
house.head()
```

**Distribution plot**

```python
sns.distplot(house['HouseAge'])
```

**Correlation matrix**  
*  1.positive correlation
*  2.Negative correlation

**Heat Map**

```python
correlation= house.corr()
```

```python
# Constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,annot=True,cbar=True,square=True,fmt='.1f',annot_kws={'size':8},cmap='Blues')
```
